```python
import os
import pandas as pd
import numpy as np
import natsort
pd.set_option('display.max_rows', None)
from decimal import *
getcontext().prec = 2

import math
```

```python
file = pd.read_excel('../../test/Purchase Invoice.xlsx',sheet_name  = 'Lines',skiprows=1)
df = pd.DataFrame(file)
df['Receipt Code'] = ['Empty'] * len(df)
i = 0
while i < len(df):
    sub_index = i+1

    if df['Description'][i].startswith('Receipt No.'):
        # this is a starting point of a receipt
        temp_receipt = df['Description'][i][12:-1]
        df['Receipt Code'][i] = ''

        while df['Description'][sub_index].startswith('Receipt No.')==False:
            df['Receipt Code'][sub_index] = temp_receipt
            sub_index = sub_index+1
            if sub_index >= len(df):
                break
    i = sub_index


df.to_excel('../../../Desktop/test/Purchase Invoice Added.xlsx', header=True,encoding='utf_8_sig')
```
