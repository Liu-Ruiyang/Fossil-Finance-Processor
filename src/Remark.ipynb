```python
def read_ali(file_name):
    file = pd.read_csv(open(file_name),skiprows=4,header=0,index_col=0,
        names=['账务流水号','业务流水号','商户订单号','商品名称','发生时间','对方账号','收入金额',
      '支出金额（-元）','账户余额（元）','交易渠道',
      '业务类型','备注','业务描述','业务账单来源','业务基础订单号','业务订单号'])
    # drop all data when ID is nan

    file = file.dropna(subset = ['业务基础订单号'])
    df = pd.DataFrame(file)
    df['业务基础订单号'] = df['业务基础订单号'].str.strip('\t')
    df_short = df[['业务基础订单号','业务流水号','商户订单号','商品名称','业务类型','备注']]
    df_short.set_index(["业务基础订单号"])
    return df_short
```

```python
def category(df):
    array = np.array(df)
    z = np.zeros((len(array),1), dtype=str)
    array = np.hstack((array,z))
    for i in range (0,len(array)):
        if('退款' in str(array[i,4]) or '在线支付' in str(array[i,4])):
            array[i,6] = 'revenue'
        else:
            if('公益宝贝' in str(array[i,5])):
                array[i,6] = 'donation'
            elif('技术年费' in str(array[i,5])):
                array[i,6] = 'annual fee'
            elif('天猫佣金' in str(array[i,5]) or '价格保障' in str(array[i,5])):
                array[i,6] = 'commission'
            elif('村淘' in str(array[i,5])):
                array[i,6] = 'commission-CT'
            elif('花呗' in str(array[i,5])):
                array[i,6] = 'commission-HB'
            elif('聚划算' in str(array[i,5]) and '聚划算保证金' not in str(array[i,5]) and '聚划算运费险' not in str(array[i,5])):
                array[i,6] = 'commission-JHS'
            elif('聚划算保证金' in str(array[i,5])):
                array[i,6] = 'commission-JHS-Deposit'
            elif('淘宝客' in str(array[i,5])):
                array[i,6] = 'commission-TBK'
            elif('价格保障' in str(array[i,5])):
                array[i,6] = 'commission-价格保障'
            elif('信用卡' in str(array[i,5])):
                array[i,6] = 'credit card'
            elif('卖家版运费险保费' in str(array[i,5])):
                array[i,6] = 'freight'
            elif('聚划算运费险' in str(array[i,5])):
                array[i,6] = 'donation'
            elif('软件服务费' in str(array[i,5]) or '服务费' in str(array[i,5])):
                array[i,6] = 'Software fee'
            elif('充值' in str(array[i,5])):
                array[i,6] = '充值'
            elif('红包' in str(array[i,5])):
                array[i,6] = '红包'
            elif('积分' in str(array[i,5])):
                array[i,6] = '积分'
            elif('提现' in str(array[i,5])):
                array[i,6] = '提现'
            else:
                print('Unknown type in Category!')
                print('type is ' + str(array[i,5]) + ', ID is '+ str(array[i,1]))
    dataset = pd.DataFrame({'账务流水号': array[:,0]+'\t',
                            '业务流水号':array[:,1],
                            '商户订单号':array[:,2],
                            '商品名称':array[:,3],
                            '业务类型': array[:,4], 
                            '备注': array[:, 5],
                            'Remark' : array[:,6]})
    return dataset
```

```python
df = read_ali('../../../Desktop/test/8-12/202008_2088311966171753.zip_files/20883119661717530156_202008_╒╦╬±├≈╧╕_1.csv')
df = category(df)
df.to_csv('../../../Desktop/test/8-12/category_test4.csv', sep=',', header=True, index=True,encoding='utf_8_sig')
```
