```python
import os
import pandas as pd
import numpy as np
import natsort
pd.set_option('display.max_rows', None)
from decimal import *
getcontext().prec = 2

import math
```

```python
def process_nav(file_name):
    # only remain when count is less than 0, which stands for payment
    # count each payment according to Document No.
    file = pd.read_excel(file_name,sheet_name  = 'Sheet1')
    df = pd.DataFrame(file)
    df = df.sort_values(by=['Document No.'],ignore_index = True)
    df['Sum of Payment'] = ['{:20,.2f}'.format(0)] * len(df)
    i = 0
    while i in range (0,len(df)):
        if df['Amount'][i] >= 0:
            i = i+1
            continue
        else:
            # amount < 0, need to count sum
            temp_sum = df['Amount'][i]
            j = i+1
            while j in range(i+1,len(df)) and  df['Amount'][j]<0 and (df['Document No.'][i]==df['Document No.'][j]):
                temp_sum = temp_sum + df['Amount'][j]
                j = j+1
            df['Sum of Payment'][i] = temp_sum
            i = j
    return df
```

```python
'''
    main
'''
df = process_nav('../../test/P1 NAV.xlsx')
df.to_excel('../../../Desktop/test/Joy Check.xlsx', header=True,encoding='utf_8_sig')
```
